```python
from nltk.stem import PorterStemmer
pst = PorterStemmer()
```

```python
pst.stem('affection')
```

```python
w = ['playing', 'maximum', 'loving', 'giving', 'thinking', 'giving', 'saved', 'Forgotten', 'died']

for i in w:
    print(i+':',pst.stem(i))
```

```python
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()

for i in w:
    print(i+':',lst.stem(i))
```

```python
from nltk.stem import SnowballStemmer
sbst = SnowballStemmer('english')

for i in w:
    print(i+':',sbst.stem(i))
```

```python
from nltk.stem import wordnet
from nltk.stem import WordNetLemmatizer

word_lem = WordNetLemmatizer()
```

```python
for i in w:
    print(i+':',WordNetLemmatizer.lemmatize(i))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Sample professional text
text = """
Artificial Intelligence (AI) and Machine Learning (ML) are transforming industries globally. 
With advancements in Natural Language Processing (NLP), organizations are leveraging chatbots and virtual assistants to enhance customer engagement.
Moreover, predictive analytics powered by AI is enabling businesses to make data-driven decisions, optimize operations, and forecast trends effectively.
Cloud computing and edge technologies further augment AI adoption by providing scalable and flexible infrastructure.
The ethical implications of AI, such as fairness, accountability, and transparency, remain critical challenges that require collaborative efforts across academia, industry, and policymakers.
"""

# Generate the WordCloud with dark theme settings
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='black',  # Dark background
    colormap='cool',           # Cool, vibrant color palette
    contour_width=2,           # Border around the WordCloud
    contour_color='steelblue', # Border color to match the dark theme
    max_font_size=100,
    min_font_size=10,
).generate(text)

# Display the WordCloud
plt.figure(figsize=(9, 4), facecolor='black')  # Dark background for the figure
plt.imshow(wordcloud, interpolation='bilinear')  # Smooth rendering
plt.axis("off")  # Turn off axis
plt.margins(x=0, y=0)  # No margins
# plt.title("Dark Aesthetic WordCloud", fontsize=18, color='white', weight='bold')  # Add an aesthetic title
plt.show()
```
